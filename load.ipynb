```python
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
import pandas as pd
```

```python
df = pd.read_csv('Output/monthly.csv')
time = df['Date']
pie = pd.read_csv('Output/pie.csv')
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig, ax = plt.subplots()
plt.rcParams['lines.linewidth'] = 0.5

ax.plot(time, 'Telegram', data=df, label="Telegram", marker='o')
ax.plot(time, 'Exercise_and_Samsung_notes', data=df, label="Exercise", marker='o')
ax.plot(time, 'Chrome', data=df, label="Chrome", marker='o')
ax.plot(time, 'ReadEra', data=df, label="ReadEra", marker='o')

plt.gcf().autofmt_xdate()

ax.set_title("Average daily usage per day")
ax.set_xlabel("Months")
ax.set_ylabel("Minutes")

ax.legend(loc="upper right")

plt.tight_layout()
plt.savefig("Image/average.png")
plt.show()
```

```python
fig, ax = plt.subplots()
plt.rcParams['lines.linewidth'] = 0.5

ax.plot(time, 'Total', data=df, label="Total", marker='o')

plt.gcf().autofmt_xdate()

ax.set_title("Average daily usage per day")
ax.set_xlabel("Months")
ax.set_ylabel("Minutes")

ax.legend(loc="upper right")

plt.tight_layout()
plt.savefig("Image/total.png")
plt.show()
```

```python
fig, ax = plt.subplots()
pie.rename(columns={'Exercise_and_Samsung_notes': 'Exercise'}, inplace=True)
labels = list(pie)[:-1]
total = pie.iloc[0, 5]
data = pie.iloc[0, 0:5].tolist()
explode = [0.1, 0, 0, 0, 0]
ax.pie(
    data,
    labels=labels,
    wedgeprops={"edgecolor": "black"},
    explode=explode,
    shadow=True,
    autopct="%1.1f%%",
    data=pie
)

ax.text(1.5, 1, f'Total: {total} minutes', horizontalalignment='right',
     verticalalignment='top', transform=ax.transAxes, bbox=dict(facecolor='red', alpha=0.5))
ax.set_title("Pie Chart")
plt.savefig('Image/pie.png')
plt.tight_layout()
plt.show()
```
